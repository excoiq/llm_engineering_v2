```python
import os
from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI
openai = OpenAI()
```

```python
# Load environment variables in a file called .env

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python
# System prompt

system_prompt = """
You are a professional Business Entity Extractor. 
Your sole task is to meticulously scan the provided article and return a comprehensive, distinct list of every company, corporation, or business entity mentioned. The output must be a simple, bulleted list of names.
"""
```

```python
# User prompt

user_prompt = """
    Analyze the following article text and provide the resulting list of unique business entities

"""
```

```python
def businesses_from(article):
    return [
        {"role" : "system" , "content" : system_prompt},
        {"role" : "user" , "content" : user_prompt + article}
    ]
```

```python
def businesses_mentioned(url):
    article = fetch_website_contents(url)
    response = openai.chat.completions.create(
        model = "gpt-4.1-mini",
        messages = businesses_from(article)
    )
    return response.choices[0].message.content
```

```python
def list_of_businesses(url):
    list = businesses_mentioned(url)
    display(Markdown(list))
    
```

```python
list_of_businesses("https://www.adexchanger.com/tv/its-getting-a-bit-easier-to-find-amazon-shoppers-on-samsung-tvs/")
```
